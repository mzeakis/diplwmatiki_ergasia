```python
import numpy
import skimage.io
import os
import cv2

# load audio. Using example from librosa

path = "C:/outs/spects4096"
os.chdir(path)
audio_files = os.listdir()

keys = ['C','C#','Db','D','D#','Eb','E','F','F#','Gb','G','G#','Ab','A','A#','Bb','B']

#ann_file = open("C:/Users/Michalis Zeakis/Desktop/university/ptyxiaki/Annotations.txt", "r")
#new_ann = open("C:/Users/Michalis Zeakis/Desktop/university/ptyxiaki/keyFinderNN/croppedAnnotations.txt", "a")

ann_file = open("C:/outs/shiftedAnnotations4096.txt", "r")
new_ann = open("C:/outs/square4096.txt", "a+")

while True:
    line = ann_file.readline()
    
    if not line:
        break;
    
    f_id, key, chord = line.split(' ', 2)
    #print(key)
    f_id = f_id.strip()
    key = key.strip()
    chord = chord.strip()
    
    for i in keys:

        if (key == i):
            file = f_id + ".LOFI.png"
            spect = os.path.join(path, file)
            spect = cv2.imread(spect, cv2.IMREAD_UNCHANGED)
            crop1 = spect[0:105, 0:105]
            crop2 = spect[0:105, 105:210] 
            crop3 = spect[0:105, 210:315]
            crop4 = spect[0:105, 315:420]
            crop5 = spect[0:105, 420:525]
            #crop1 = cv2.resize(crop1, (600, 105))        #croping image to half and resize to (105, 600)
            #crop2 = cv2.resize(crop2, (600, 105))
            
            new_name = f_id + "a"
            new_ann.write(new_name)                  #annotations for crop1
            new_ann.write(" ")
            new_ann.write(key)
            new_ann.write(" ")
            new_ann.write(chord)
            new_ann.write("\n")
            new_name = new_name + ".LOFI"
            out = 'C:/outs/square4096/{0}.png'.format(new_name)   #saving spect for crop1
            skimage.io.imsave(out, crop1)
            
            new_name = f_id + "b"
            new_ann.write(new_name)
            new_ann.write(" ")
            new_ann.write(key)                      #annotations for crop2
            new_ann.write(" ")
            new_ann.write(chord)
            new_ann.write("\n")
            new_name = new_name + ".LOFI"
            out = 'C:/outs/square4096/{0}.png'.format(new_name)   #saving spect for crop2
            skimage.io.imsave(out, crop2)
            
            new_name = f_id + "c"
            new_ann.write(new_name)
            new_ann.write(" ")
            new_ann.write(key)                      #annotations for crop2
            new_ann.write(" ")
            new_ann.write(chord)
            new_ann.write("\n")
            new_name = new_name + ".LOFI"
            out = 'C:/outs/square4096/{0}.png'.format(new_name)   #saving spect for crop3
            skimage.io.imsave(out, crop3)
            
            new_name = f_id + "d"
            new_ann.write(new_name)
            new_ann.write(" ")
            new_ann.write(key)                      #annotations for crop2
            new_ann.write(" ")
            new_ann.write(chord)
            new_ann.write("\n")
            new_name = new_name + ".LOFI"
            out = 'C:/outs/square4096/{0}.png'.format(new_name)   #saving spect for crop4
            skimage.io.imsave(out, crop4)
            
            new_name = f_id + "e"
            new_ann.write(new_name)
            new_ann.write(" ")
            new_ann.write(key)                      #annotations for crop2
            new_ann.write(" ")
            new_ann.write(chord)
            new_ann.write("\n")
            new_name = new_name + ".LOFI"
            out = 'C:/outs/square4096/{0}.png'.format(new_name)   #saving spect for crop5
            skimage.io.imsave(out, crop5)
            

ann_file.close()
new_ann.close()

print("finito")
```
